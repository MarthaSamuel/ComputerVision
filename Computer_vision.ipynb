# image segmentation

```python
img = cv2.imread('coin.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# PIXEL>0=255 ELSE 0
# PIXEL>0=0 ELSE 255
ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

kernel = np.ones((3,3), np.uint8)


#noise removal
# erosion followed by dilation :noise
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel, iterations=3)

#background area
sure_bg = cv2.dilate(opening, kernel, iterations=3)

#replaced with nearest bg values
dts = cv2.distanceTransform(opening, cv2.DIST_L2, maskSize=5)#finds fg area
ret, sure_fg = cv2.threshold(dts, 0.7*dts.max(), 255, 0)

sure_fg = np.uint8(sure_fg)

img2 = cv2.subtract(sure_bg, sure_fg)
ret , markers = cv2.connectedComponents(sure_fg) #this labels marks.bg is 0, 1st
#img in bg is 1,2nd img in bg is 2 etc

markers[img2==255]= 0
markers = cv2.watershed(img,markers )#watershed is used for image segmentationout

img[markers==-1]=[255,255,0]

plt.imshow(img)
```
